```
I1 = cv2.imread('./images/boat.png')
I2 = cv2.imread('./images/cameraman.png')
fig = plt.figure(figsize=(12,8))
J1= cv2.add(I1, I2)
J2 = cv2.subtract(I1, I2)
J3 = cv2.absdiff(I1, I2)
J4 = cv2.multiply(I1, I2)
J5 = cv2.divide(I1, I2)
J6 = cv2.addWeighted(I1,0.7,I2,0.3,0)
J7 = cv2.bitwise_and(I1, I2)
J8 = cv2.bitwise_or(I1, I2)
plt.subplot(521),plt.imshow(I1), plt.axis('off')
plt.subplot(522),plt.imshow(I2), plt.axis('off')
plt.subplot(523),plt.imshow(J1), plt.axis('off')
plt.subplot(524),plt.imshow(J2), plt.axis('off')
plt.subplot(525),plt.imshow(J3), plt.axis('off')
plt.subplot(526),plt.imshow(J4), plt.axis('off')
plt.subplot(527),plt.imshow(J5), plt.axis('off')
plt.subplot(528),plt.imshow(J6), plt.axis('off')
plt.subplot(529),plt.imshow(J7), plt.axis('off')
fig.add_subplot(5, 2, 10),plt.imshow(J8), plt.axis('off')

plt.tight_layout(pad=.2)
plt.show()
```
